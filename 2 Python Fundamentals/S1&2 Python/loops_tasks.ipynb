<h2 style="color:#00bfff;">üìå Task 1 ‚Äî Loop Control Statements</h2>

<h3 style="color:#87cefa;">pass</h3>

<span style="color:#dcdcdc;">
‚Ä¢ Does absolutely nothing  
‚Ä¢ Execution continues normally  
‚Ä¢ Used as a placeholder to avoid syntax errors  
</span>

<b style="color:#add8e6;">Key idea</b>  
`pass` changes nothing in the loop behavior.

---

<h3 style="color:#87cefa;">continue</h3>

<span style="color:#dcdcdc;">
‚Ä¢ Skips the current iteration  
‚Ä¢ Jumps directly to the next iteration  
‚Ä¢ Remaining code inside the loop is ignored for this cycle  
</span>

<b style="color:#add8e6;">Key idea</b>  
`continue` skips forward.

---

<h3 style="color:#87cefa;">break</h3>

<span style="color:#dcdcdc;">
‚Ä¢ Terminates the loop completely  
‚Ä¢ Control moves outside the loop  
</span>

<b style="color:#add8e6;">Key idea</b>  
`break` exits the loop.

---

<h4 style="color:#00ced1;">Quick Comparison</h4>

| Statement | Effect |
|---------|--------|
| pass | No effect |
| continue | Skip current iteration |
| break | Exit loop |

<h2 style="color:#00bfff;">üìå Task 2 ‚Äî for Loop vs while Loop</h2>

<h3 style="color:#87cefa;">for loop</h3>

<span style="color:#dcdcdc;">
‚Ä¢ Used when the number of iterations is known  
‚Ä¢ Iterates over sequences like range, list, or string  
</span>

<b style="color:#add8e6;">Common use cases</b>  
‚Ä¢ Fixed number of repetitions  
‚Ä¢ Iterating through lists  
‚Ä¢ Processing strings or arrays  

---

<h3 style="color:#87cefa;">while loop</h3>

<span style="color:#dcdcdc;">
‚Ä¢ Used when the number of iterations is unknown  
‚Ä¢ Runs as long as a condition remains True  
</span>

<b style="color:#add8e6;">Common use cases</b>  
‚Ä¢ Waiting for user input  
‚Ä¢ Monitoring conditions  
‚Ä¢ Infinite loops with controlled exits  

---

<h4 style="color:#00ced1;">Rule of Thumb</h4>

<span style="color:#dcdcdc;">
Known iterations ‚Üí <b>for</b>  
Condition based looping ‚Üí <b>while</b>  
</span>

```python
def stop_at_five_or_thirty():
    for number in range(1, 100):
        if number == 5 or number == 30:
            break
        print(number)


````python
def loop_except_5_or_30():
    for num in range(0, 41):
        if num == 5 or num == 30:
            continue
        print(num, end=", ")
````

Task 1
Difference between pass, continue, and break inside a for loop
These three control how a loop behaves when it hits a certain condition.
pass
Does nothing. Literally nothing.
The loop continues as if the line does not exist.
It is a placeholder when syntax requires a statement.
Use case
You are sketching logic.
You want to fill it later.
You want the loop to keep running normally.
continue
Skips the current iteration only.
The loop jumps directly to the next iteration.
Use case
You want to ignore certain values.
You do not want the rest of the loop body to run for this iteration.
break
Stops the loop completely.
Execution jumps outside the loop.
Use case
You found what you were looking for.
There is no reason to keep looping.
Key mental model
pass does nothing
continue skips forward
break exits entirely

Task 2
Difference between a for loop and a while loop
for loop
Used when you know the number of iterations in advance.
It iterates over a sequence like a list, range, or string.
Think
You know the start and the end.
You are walking through something finite.
Example use cases
Looping through numbers 1 to 10
Iterating over elements in a list
Processing characters in a string
while loop
Used when the number of iterations is unknown.
It runs as long as a condition is true.
Think
You are waiting for something to happen.
You stop only when a condition changes.
Example use cases
Waiting for user input
Running until a value reaches a threshold
Reading data until a file ends
Rule of thumb
Known count ‚Üí for
Unknown duration ‚Üí while
