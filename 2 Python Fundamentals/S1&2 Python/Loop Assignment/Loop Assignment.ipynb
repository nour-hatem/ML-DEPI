2. Go to the data folder and use the **[countries_details_data.py]()** file. 
   1. What are the total number of languages in the data \\ 112
   2. Find the ten most spoken languages from the data \\ English , French , Arabic
   3. Find the 10 most populated countries in the world \\ China, India, US

````python
total_langs = set()

for country in countries_data:
    for lang in country["languages"]:
        total_langs.add(lang)

len(total_langs)
````

````python
lang_count = {}

for country in countries_data:
    for lang in country["languages"]:
        if lang in lang_count:
            lang_count[lang] += 1
        else:
            lang_count[lang] = 1

top_10 = []

for x in range(10):
    max_lang = None
    max_count = -1

    for lang in lang_count:
        if lang_count[lang] > max_count:
            max_count = lang_count[lang]
            max_lang = lang

    top_10.append(max_lang)
    del lang_count[max_lang]

for lang in top_10:
    print(lang)
````

````python
top_10_pop = []
countries_copy = countries_data.copy()

for x in range(10):
    max_country = None
    max_pop = -1

    for country in countries_copy:
        if country["population"] > max_pop:
            max_pop = country["population"]
            max_country = country

    top_10_pop.append(max_country["name"])
    countries_copy.remove(max_country)

for country in top_10_pop:
    print(country)
````
