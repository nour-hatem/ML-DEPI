```python
from abc import ABC , abstractmethod


class PaymentMethod(ABC): #1) Abstraction 
    @abstractmethod
    def pay(self, amount):
        pass
    
    @abstractmethod
    def deposit(self, amount):
        pass
```

```python
from abc import ABC , abstractmethod


class PaymentMethod(ABC): #1) Abstraction 
    @abstractmethod
    def pay(self, amount):
        pass
```

```python
class Account: 
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  #  Encapsulation
    # deposit, withdarw , get_balance 
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit must be positive")
        self.__balance += amount
        
    def withdarw(self, amount):
        if amount <= 0:
            raise ValueError("withdarw must be positive")
        if amount > self.__balance:
            raise ValueError("U don't have balance")
        self.__balance -= amount

    def get_balance(self):
        return self.__balance
    
    
```

```python
class card_Payment(PaymentMethod): #2) Inheritance:  card_Payment inherit form PaymentMethod ABC class
    def __init__(self,account,last4Num):
        self.account =account
        self.last4Num =last4Num
    
    def pay(self,amount):
        self.account.withdarw(amount)
        return f"Paid ${amount} using card *****{self.last4Num}"
```

```python
class instaPay_Payment(PaymentMethod): #2) Inheritance:  card_Payment inherit form PaymentMethod ABC class
    def __init__(self,account,IpnUrl):
        self.account =account
        self.IpnUrl =IpnUrl
    
    def pay(self,amount):
        self.account.withdarw(amount)
        return f"Paid ${amount} using user {self.IpnUrl}"
```

```python
def checkout(paymentMethod, amount):
    print(paymentMethod.pay(amount))
```

```python
acc = Account("Baraa",100)
card= card_Payment(acc,1234)
instaPay= instaPay_Payment(acc,"engbaraa@instapay")


checkout(instaPay,30)
checkout(card,30)
```
