```python
class Person:
    startup = "AMIT"
    def __init__(self, name, age):
        self.nameEmp = name
        self.__ageEmp = age
        
    def greet(self):
        print(f"Hello {self.nameEmp}, {self.__ageEmp} in our {self.startup}")
    
    def get_age(self):
        return self.__ageEmp

    def set_age(self, new_age):
        self.__ageEmp = new_age
    
    def dance(self, name):
        print(f"{name} Dance")

emp1 = Person("Nour", 20)
emp1.greet()
print("Before:", emp1.get_age())
print("Edit:", emp1.set_age(80))
print("After:", emp1.get_age())
emp1.greet()
```

```python
class my_statistics:
    def __init__(self, list):
        self.list = list

    def count(self):
        return len(self.list)

    def sum(self):
        return sum(self.list)

    def mean(self):
        return sum(self.list) / len(self.list)

    def median(self):
        sorted_list = sorted(self.list)
        n = len(sorted_list)
        
        mid2 = sorted_list[n // 2]
        if n % 2 == 0:
            mid1 = sorted_list[n // 2 - 1]
            return (mid1 + mid2) / 2
        return mid2


ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27]
Calc = my_statistics(ages)

print(Calc.count())
print(Calc.sum())
print(Calc.mean())
print(Calc.median())
```

```python
class VotingSystem:
    def __init__(self):
        self.candidates = {}

    def add_candidate(self, name):
        if name in self.candidates:
            print(f"Candidate {name} already exists.")
        else:
            self.candidates[name] = 0

    def show_candidates(self):
        if not self.candidates:
            print("No candidates available.")
        else:
            print(list(self.candidates.keys()))

    def vote(self, name):
        if name not in self.candidates:
            print(f"Candidate {name} isnot on the list.")
        else:
            self.candidates[name] += 1

    def get_results(self, name):
        if name not in self.candidates:
            print("No such candidate.")
        else:
            print(f"Number of votes for {name} are: {self.candidates[name]}")

    def display_winner(self):
        if not self.candidates:
            print("No candidates. No winner.")
            return

        winner = max(self.candidates, key=self.candidates.get)
        votes = self.candidates[winner]
        print(f"The winner is {winner} with {votes} votes. ^_^")
```

```python
voting = VotingSystem()

voting.add_candidate("Ali")
voting.add_candidate("Nour")
voting.add_candidate("Mohamed")

voting.show_candidates()

voting.vote("Ali")
voting.vote("Nour")
voting.vote("Nour")

voting.vote("Nour")
voting.vote("Mohamed")

voting.vote("sss")

voting.get_results("Nour")
voting.get_results("Ali")
voting.get_results("Mohamed")
voting.get_results("sss")

voting.display_winner()
```
